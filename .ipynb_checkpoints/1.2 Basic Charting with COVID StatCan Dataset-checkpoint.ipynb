```python

import pandas as pd
import plotly.graph_objects as go
import numpy as np
from stats_can import StatsCan
sc = StatsCan()

#df = sc.table_to_df("271-000-22-01")
#df = sc.vectors_to_df(["v74804", "v41692457"])
df = sc.table_to_df("131-007-66-01")

pd.set_option("display.max_rows", None, "display.max_columns", None)
df.head()
```

```python
to_drop = ['GEO', 
 'DGUID', 
 'UOM', 
 'SCALAR_FACTOR', 
 'STATUS', 
 'SYMBOL', 
 'TERMINATED', 
 'DECIMALS']

df.drop(columns=to_drop, inplace=True)
```

```python
df.head()
```

```python
df2 = df.pivot(index='Case identifier number', columns='Case information', values='VALUE')
```

```python
df2.head()
```

```python
#removing unanswered age group rows for display
age_df = df2.drop(df2[(df2['Age group'] == 99.0)].index)
age_df_m = age_df.drop(age_df[(age_df['Sex'] == 1.0)].index) # male
age_df_f = age_df.drop(age_df[(age_df['Sex'] == 2.0)].index) # female
age_df_u = age_df.drop(age_df[(age_df['Sex'] == 7.0)].index) # unknown

fig = go.Figure()
fig.add_trace(go.Histogram(
    x=age_df_m['Age group'],
    name = 'male'
))

fig.add_trace(go.Histogram(
    x=age_df_f['Age group'],
    name = 'female'
))

fig.add_trace(go.Histogram(
    x=age_df_u['Age group'],
    name = 'unknown'
))

fig.update_layout(
    title_text='Age Distribution of Canadian COVID-19 Cases', # title of plot
    xaxis_title_text='Age Group', # xaxis label
    yaxis_title_text='Count', # yaxis label
    #bargap=0.1, # gap between bars of adjacent location coordinates
    #bargroupgap=0.1, # gap between bars of the same location coordinates


)

fig.update_traces(opacity=0.75)

fig.show()
```

Age group codes: 
1 = 0 to 19 years, 
2 = 20 to 29 years, 
3 = 30 to 39 years, 
4 = 40 to 49 years, 
5 = 50 to 59 years, 
6 = 60 to 69 years, 
7 = 70 to 79 years, 
8 = 80 years or older,
